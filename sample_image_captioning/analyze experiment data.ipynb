```python
# average classes with and without caption
person_data = pd.read_csv(f"experiments/exp_mask-4-main_thr-0-sec_thr-0/class_scores.csv")
person_data["model_name"] = "exp_mask-4-main_thr-0-sec_thr-0"
person_data["avg_caption_objects"] = 0
person_data["avg_no_caption_objects"] = 0
person_data["diff"] = 0
for idx, row in person_data.iterrows():
    avg_no_caption = np.mean([row["backpack"], row["handbag"]])
    avg_caption = np.mean([row["suitcase"], row["tie"], row["umbrella"]])
    person_data.at[idx, "avg_caption_objects"] = avg_caption
    person_data.at[idx, "avg_no_caption_objects"] = avg_no_caption
    person_data.at[idx, "diff"] = avg_caption - avg_no_caption
person_data[person_data["object"]=="main"][["model_name", "object", "token_strategy", "hidden_state_layer", "avg_caption_objects", "avg_no_caption_objects", "diff"]]
```

```python
# same as above but all classes without caption
# average classes with and without caption
person_data = pd.read_csv(f"experiments/exp_mask-4-main_thr-0-sec_thr-0_no_caption/class_scores.csv")
person_data["model_name"] = "exp_mask-4-main_thr-0-sec_thr-0_no_caption"
person_data["avg_caption_objects"] = 0
person_data["avg_no_caption_objects"] = 0
person_data["diff"] = 0
for idx, row in person_data.iterrows():
    avg_no_caption = np.mean([row["backpack"], row["handbag"]])
    avg_caption = np.mean([row["suitcase"], row["tie"], row["umbrella"]])
    person_data.at[idx, "avg_caption_objects"] = avg_caption
    person_data.at[idx, "avg_no_caption_objects"] = avg_no_caption
    person_data.at[idx, "diff"] = avg_caption - avg_no_caption
person_data[person_data["object"]=="main"][["model_name", "object", "token_strategy", "hidden_state_layer", "avg_caption_objects", "avg_no_caption_objects", "diff"]]
```
